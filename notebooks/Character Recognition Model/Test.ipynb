```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/Colab Notebooks/Stage 2/run_n/exp/weights/last.pt")
metrics = model.val(
    data='/content/drive/MyDrive/Colab Notebooks/Stage 2/dataset.yaml',
    split="test",
    imgsz=640,
    batch=100,
    device="cuda",
    verbose=True
)

print(metrics.box.map)
print(metrics.box.map50)
print(metrics.speed)
```
